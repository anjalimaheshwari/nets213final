```python
from gtts import gTTS 
from random import randint
from pydub import AudioSegment
from midi2audio import FluidSynth
from IPython.display import Audio
import pandas as pd
```

```python
# pick sample music
midi1 = "lmd_full/0/00000ec8a66b6bd2ef809b0443eeae41.mid"
midi2 = "lmd_full/0/0000799e8672292fe6f0fff08554ca40.mid"
midi3 = "lmd_full/0/000a30cffe45faaca2d0b919d900961c.mid"
sample_input = [midi1, midi2, midi3]
```

```python
def getSpeech():
  num = randint(10, 99)
  text = "Please enter" + str(num) + "into the textbox"
  language = 'en'
  speech = gTTS(text = text, lang = language, slow = False)
  return num, speech

def combineMusicSpeech (midi_fp):
  FluidSynth("font.sf2").midi_to_audio(midi_fp, 'music.wav')
  num, speech = getSpeech()
  speech.save('text.mp3')
  sound1 = AudioSegment.from_file('music.wav', format="wav")
  sound2 = AudioSegment.from_file("text.mp3", format="mp3")
  combined = sound1 + sound2
  return num, combined
  
```

```python
outfile = []
for i, midi_fp in enumerate(sample_input):
  out_fp = "outdata/combined" + str(i) + ".mp3"
  num, combined = combineMusicSpeech(midi_fp)
  combined.export(out_fp, format="mp3")
  outfile.append((out_fp, num))
```

```python
df = pd.DataFrame(outfile, columns = ["filepath", "number"])
df.to_csv("output_file.csv", index=False)
```
